```python
from sklearn.cross_validation import cross_val_score #K折交叉验证模块
```

```python
from sklearn.model_selection import train_test_split


X=train
y=label
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.9, random_state=12)
```

```python
'''
划分测试验证集来训练
'''

clf = lgbm.LGBMClassifier(
        boosting_type='gbdt', num_leaves=15, reg_alpha=0.0, reg_lambda=5,
        max_depth=-1, n_estimators=100, objective='binary',
        subsample=0.6, colsample_bytree=0.7, subsample_freq=1,
        learning_rate=0.05, min_child_weight=50,random_state=2018,n_jobs=-1
    )
clf.fit(train_x, train_y,eval_set=[(train_x, train_y),(test_x,test_y)],eval_metric='auc',early_stopping_rounds=100)
# print(clf.feature_importances_)
print(clf,clf.best_score_[ 'valid_1']['auc'])
```

```python
'''
预处理label数据，以防cv出现bug
'''

c, r = label.shape
label = label.values.reshape(c,)
```

```python
'''
用普通的lr
'''

from sklearn.linear_model import LogisticRegression

clf2 = LogisticRegression()
# clf2.fit(train,label)

scores = cross_val_score(clf2, train, label, cv = 5, scoring = 'roc_auc')
# scores = cross_val_score(clf, train, label, cv = 5, scoring = 'f1')
# 输出5次交叉验证的准确率
print (scores)
```

```python
'''
用随机森林
'''

from sklearn.ensemble import RandomForestClassifier

clf3 = RandomForestClassifier()
# clf2.fit(train,label)

scores = cross_val_score(clf3, train, label, cv = 5, scoring = 'roc_auc')
# scores = cross_val_score(clf, train, label, cv = 5, scoring = 'f1')
# 输出5次交叉验证的准确率
print (scores)
```
